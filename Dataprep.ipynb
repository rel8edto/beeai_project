```python
#!/usr/bin/env python3
import os, json, sys
from openai import OpenAI

def main(company: str = "onerep") -> None:
    key = os.getenv("OCTAGON_API_KEY")
    if not key:
        print("❌  Set OCTAGON_API_KEY first."); return

    client = OpenAI(api_key=key,
                    base_url="https://api-gateway.octagonagents.com/v1")

    # ── 1. ticker lookup ────────────────────────────────────────────────
    query = (f"Return ONLY the primary stock-ticker symbol for '{company}'. "
             f"If it is not publicly traded, reply 'PRIVATE'.")
    sym_resp = client.responses.create(model="octagon-stock-data-agent",
                                       input=query)

    ticker = "".join(p.text for p in sym_resp.output[0].content).strip().upper()

    if ticker in {"PRIVATE", "N/A", ""}:
        print(f"🔎 {company}: company isn’t publicly listed.")
        return                          # ← clean exit; no SystemExit

    print(f"✅ Ticker symbol for {company}: {ticker}")

    # ── 2. holdings data ───────────────────────────────────────────────
    h_query = (f"Get a summary of institutional positions for {ticker} "
               f"for Q4 2024 (current) and Q3 2024 (previous). Respond in JSON.")
    h_resp  = client.responses.create(model="octagon-holdings-agent",
                                      input=h_query)

    raw = "".join(p.text for p in h_resp.output[0].content).strip()
    try:
        data = json.loads(raw)
    except json.JSONDecodeError:
        print("⚠️  Octagon did not return JSON; raw reply below:\n", raw)
        return

    print(json.dumps(data, indent=2))

if __name__ == "__main__":
    main("onerep")          # or pass sys.argv[1] for CLI usage
```
